**Больше отрицательных прогнозов / точность будет выше**. Это означает, что при повышении порога мы будем делать меньше ошибок второго рода (ложноположительные), но больше ошибок первого рода (ложноотрицательные). Это может быть желательно или нежелательно в зависимости от цели и контекста задачи. Например, в медицине мы можем предпочесть более высокую чувствительность, чтобы не пропустить больных пациентов, а в безопасности мы можем предпочесть более высокую точность, чтобы не ложно обвинять невиновных людей.
