```python
import pandas as pd
import requests
```

```python
url_link = 'https://finance.yahoo.com/quote/%5EGSPC/history?period1=1641166055&period2=1672702055&interval=1wk&filter=history&frequency=1wk&includeAdjustedClose=true'
r = requests.get(url_link,headers ={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'})

data = pd.read_html(r.text)
```
