Finding stocks that have low correlation helps to diversify an investment portfolio and minimize risk.

#### Let's create our portfolio as 10% EPAM, 30% GOOGLE, 35% META and 25% MICROSOFT

```python
import yfinance as yf
import numpy as np
```

```python
stocks = ['EPAM', 'GOOG', 'META', 'MSFT']
weights = [0.1, 0.3, 0.35, 0.25]
```

```python
data = yf.download(stocks, start='2021-01-01')['Close']
data
```

```python
daily_returns = data.pct_change()
daily_returns
```

```python
corr = daily_returns.corr()
corr
```

```python
portfolio_return = (daily_returns * weights).sum(axis=1)
portfolio_return
portfolio_return.plot()
```

```python
total_cumulative_returns = (portfolio_return + 1).cumprod()
total_cumulative_returns
```

```python
total_cumulative_returns.plot()
# returns for each 1$ of investments
```

**Volatility** is also often used to measure risk. If a stock is very volatile, you can expect large changes in its price and therefore a higher risk.

Volatility is calculated using the standard deviation of the portfolio return.

```python
import numpy as np

np.std(total_cumulative_returns)
```

```python
annual_std = np.std(total_cumulative_returns) * np.sqrt(252)
# there are 252 traiding days in a year
print(f'risk % of our portfolio is {annual_std}')
```

Another important metric is the Sharpe ratio.
Sharpe ratio is the measure of the risk-adjusted return of a portfolio. A portfolio with a higher Sharpe ratio is considered better.

To calculate the Sharpe ratio, we need to take the average return and divide it by the volatility.

```python
sharpe = (np.mean(portfolio_return) / np.std(portfolio_return)) * np.sqrt(252)
sharpe
```

We multiplied the result by the square root of 252 to annualize the Sharpe ratio. (There are 252 trading days in a year)

Sharpe ratios greater than 1 are considered optimal.

## Portfolio optimization

We calculated the main metrics to analyze and evaluate a portfolio of stocks. Now we can use the power of Python to optimize a portfolio!

Portfolio optimization is the technique of allocating assets so that it has the maximum return and minimum risk.
This can be done by finding the allocation that results in the maximum Sharpe ratio.

The simplest way to find the best allocation is to check many random allocations and find the one that has the best Sharpe ratio.

This process of randomly guessing is known as a Monte Carlo Simulation.

```python
stocks = ['EPAM', 'META', 'AAPL', 'NVDA', 'AMD']
data = yf.download(stocks, start='2022-01-01')

# daily returns
data = data['Close']
price_changes = data.pct_change()
price_changes
```

We need to store the weights, returns and Sharpe ratios for each portfolio we will be checking so that we can then find the one that is best.
We will define lists to store these values for each portfolio.

```python
weights = []
returns = []
risk = []
sharpe = []
```

We are going to randomly assign a weight to each stock in our portfolio, and then calculate the metrics for that portfolio, including the Sharpe ratio.

```python
w = np.random.uniform(size=len(price_changes.columns))
print(w)  # random weight
w = w / np.sum(w)
print(w)  # shares
```

We divide the resulting weights by their sum to normalize them, so that the sum of the random weights is always 1.
Running the code above will generate random weights that sum up to 1.


```python
loops = 1_000
for loop in range(loops):
    # weights
    weight = np.random.uniform(size=len(price_changes.columns))
    weight = weight / np.sum(weight)
    weights.append(weight)
    
    # returns
    mean_annual_return = (price_changes.mean() * weight).sum() * 252
    returns.append(mean_annual_return)
    
    # volatility
    yields = (price_changes * weight).sum(axis=1)
    annual_std = np.std(yields) * np.sqrt(252)
    risk.append(annual_std)
    
    # Sharpe ratio
    sh_ratio = (np.mean(yields) / np.std(yields)) * np.sqrt(252)
    sharpe.append(sh_ratio)
    
```

```python
max_index = np.argmax(sharpe)
max_index # Returns the indices of the maximum values along an axis
```

```python
print(sharpe[max_index])
print(weights[max_index])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

pd.Series(weights[max_index], index=price_changes.columns).plot(kind='bar')
```

The final chart is called Efficient Frontier and shows the returns on the Y-axis and volatility on the X-axis.

We can create the chart using the scatter() function, providing the volatility and return lists as parameters:

```python
plt.scatter(risk, returns, c=sharpe, cmap='plasma')
plt.colorbar(label='Sharpe Ratio')

plt.scatter(risk[max_index], returns[max_index], color='r', marker='*', s=500)
plt.show()
```
